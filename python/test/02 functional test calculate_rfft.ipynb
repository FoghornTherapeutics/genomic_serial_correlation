```python
import numpy

import fhtbioinfpy.genomic_serial_corr.genomic_serial_corr as gensc

# import plotly.express as pltxpr
```

```python
input_cov = "/home/ec2-user/balt_wkdir/temp/H3K27ac_OCILY10_DMSO_chr1_fragments_std_norm.npy"
input_ref_result = "/home/ec2-user/balt_wkdir/118_serial_correlation/code/NS-24.0018/results/coverage/rfft_fragment_coverage_std_norm/H3K27ac_OCILY10_DMSO_chr1_fragments_fragment_cov_std_norm_rfft.npy"
```

```python
cov_arr = numpy.load(input_cov)
print(cov_arr.dtype)
print(cov_arr.shape)
cov_arr
```

```python
ref_result_arr = numpy.load(input_ref_result)
print(ref_result_arr.dtype)
print(ref_result_arr.shape)
ref_result_arr
```

```python
rfft_cov_arr = gensc.calculate_rfft(cov_arr)
print(rfft_cov_arr.dtype)
print(rfft_cov_arr.shape)
rfft_cov_arr
```

```python
assert ref_result_arr.shape == rfft_cov_arr.shape
```

```python
diff_arr = rfft_cov_arr - ref_result_arr
diff_arr
```

```python
power_spec_diff_arr = diff_arr * numpy.conj(diff_arr)
power_spec_diff_arr
```

```python
sum_power_spec_real = numpy.sum(numpy.abs(numpy.real(power_spec_diff_arr)))
print(sum_power_spec_real)
assert sum_power_spec_real < 1e-10
```

```python
sum_power_spec_imag = numpy.sum(numpy.abs(numpy.imag(power_spec_diff_arr)))
print(sum_power_spec_imag)
assert sum_power_spec_imag < 1e-10
```
